```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

series = 'Avatar-Last-Airbender'
df = pd.read_csv(f'{series}.csv')
genres = df['genre'].apply(lambda x: str(x).split('/'))
genre_counts = sorted(list(Counter(genres.sum()).items()), key=lambda x: x[1])
labels, vals = zip(*genre_counts)
y_pos = np.arange(len(labels))

plt.figure(figsize=(15, 20))
plt.barh(y_pos, vals)
plt.yticks(y_pos, labels)
plt.title("Number of Stories by Genre")
plt.xlabel("Number of stories")
plt.ylabel("Genre")
plt.show()
```
